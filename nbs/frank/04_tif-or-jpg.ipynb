# TIF or JPG?

*A question that arises now is if conversion of the TIF transmitted light images into JPG has resulted in a loss of quality.*

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
```

```python
tif_file = '/media/frank/lonprojecten/2024-811_falnama-project/data/raw/Weimar-folders/Weimar-Transmission/71803-04-Trans1.tif'

jpg_file = '/media/frank/lonprojecten/2024-811_falnama-project/viz/02. Transmitted light - JPEG/71803-04/71803-04_Trans1.jpg'
```

```python
tif_im = plt.imread(tif_file)
tif_im = tif_im.transpose(1, 0, 2)[:,::-1]
```

```python
jpg_im = plt.imread(jpg_file)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=[15, 12], sharex=True, sharey=True)
[ax0, ax1] = axs.flatten()

ax0.imshow(tif_im)
ax0.set_title('TIF')
ax1.imshow(jpg_im)
ax1.set_title('JPG');
```
